```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Cell Data Adjusted Columns.csv', index_col=0)
df = df.drop(columns=['Frame', 'Created By'])
df.head(6)
```

```python
sns.set_style("ticks") #whitegrid, darkgrid, dark, white, ticks

sns.set_context("paper") # paper, talk, poster

sns.scatterplot(x='Area', y='Extent', hue='Treatment', data=df)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),
            vmin=-1,
            cmap='coolwarm',
            annot=True)
```

```python
plt.figure(figsize=(10, 8))

plt.xticks(rotation=45)

ax = sns.boxplot(data = df,
            y = 'Area',
            x = 'Treatment')

ax.set_xlabel("Treatment", fontsize = 20)

ax.set_ylabel("Area", fontsize = 20)

ax.set_title("Box Plot", fontsize = 35)
```

```python
sns.set(style="ticks")

sns.pairplot(df, hue='Treatment')
```

```python
fig = px.histogram(df, x= "Area", color="Treatment")

fig.show()
```

```python
fig = px.histogram(df, x= "Equivalent Diameter", color="Treatment")

fig.show()
```

```python
fig = px.box(df, x="Treatment", y="Area", color="Treatment")
fig.show()
```

```python
fig = px.box(df, x="Treatment", y="Equivalent Diameter", color="Treatment")
fig.show()
```

```python
fig = px.scatter_matrix(df, 
    dimensions=["Area", "Perimeter", "Equivalent Diameter"],
    color="Treatment")
fig.show()
```

```python
plt.figure(figsize=(10,8))

sns.swarmplot(x='Treatment', y='Area', data=df, 
              hue='Treatment')

plt.axhline(df['Area'].mean(), color='red', linestyle='dashed', linewidth= 10)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
